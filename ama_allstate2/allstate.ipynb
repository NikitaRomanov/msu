```python
import numpy as np
import pandas as pd
from xgboost import XGBRegressor
from sklearn.model_selection import cross_val_score
```

```python
data = pd.read_csv(r'D:\allstate\train.csv.zip')
data.head()
```

```python
[(x, data[x].nunique()) for x in data.columns[1:-1] if 'cat' in x]
```

```python
features = data.columns
cats = [feat for feat in features if 'cat' in feat]
for feat in cats:
    data[feat] = pd.factorize(data[feat], sort=True)[0]
```

```python
data.head()
```

```python
X_train = data.drop(['id','loss'], axis=1)
y_train = data.loss
```

```python
X_train.shape
```

```python
regr = XGBRegressor()
regr.fit(X_train, y_train, verbose=1)
```

```python
cross_val_score(regr, X_train, y_train, cv=8, scoring='neg_mean_absolute_error')
```
